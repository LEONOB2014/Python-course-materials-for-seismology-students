```python
# this line is what is called "ipython magic". It allows
# inline matplotlib plots which makes for quick prototyping
# of maps, graphs, and other visuals.
%matplotlib inline

# Pandas and numpy is our main data storage library and allows us to
# manipulate our data in many ways
import pandas as pd
import numpy as np

# matplotlib allows us to create plots
import matplotlib.pyplot as plt

# cartopy allows us to create maps, ccrs holds all the available map projections
import cartopy
import cartopy.crs as ccrs
```

```python
# Here we create the ax object. It is important to do this because 
# it allows us many basic ways to manipulate the image through the 
# Axes methods
fig, ax = plt.subplots(1, figsize=(8,8))

# Here we set the map projection. It is important to NOT NOT NOT
# name this object "map". This is because map is a built-in function
# in python and you will no longer be able to use it if you call this
# object map.

proj = ccrs.PlateCarree()           

# Here we set the projection on the axes
ax = plt.axes(projection=proj)  

# Here we set the boundaries of the maps
llcrnrlat=17.5
urcrnrlat=51.5
llcrnrlon=115.0
urcrnrlon=158.0

ax.set_xlim(llcrnrlon,urcrnrlon)
ax.set_ylim(llcrnrlat,urcrnrlat)

# Here we draw some features on the map
ax.add_feature(cartopy.feature.OCEAN,zorder=0,facecolor='red')           
ax.add_feature(cartopy.feature.LAND,zorder=0,edgecolor='black',facecolor='gray')               

# This allows us to set a title for the map. We can use many methods
# from the ax object to produce labels, annotations, etc.
ax.set_title('My First Map', fontsize=20)

# For this toy example we will just plot the locations of two cities:
# Tokyo and Taipei. To do this first we need to know the longitude
# and latitude of these cities.
tokyo = [139.6917, 35.6895]

ax.plot(tokyo[0], tokyo[1],'*', color='red', markersize=15, transform=ccrs.Geodetic()) 

# Note again, the power of the Axes object.
ax.annotate('Tokyo', xy=(tokyo[0]+1, tokyo[1]), color='green', fontsize=40)

# We repeat what we did for Tokyo for Taipei
taipei = [121.5654, 25.0330]
ax.plot(taipei[0], taipei[1], '*', color='yellow', markersize=15, transform=ccrs.Geodetic())
ax.annotate('Taipei', xy=(taipei[0]+1, taipei[1]), color='yellow', fontsize=40)

# We repeat what we did for Tokyo for Beijing
beijing = [116.4074, 39.9042]
ax.plot(beijing[0], beijing[1], '*', color='blue', markersize=15, transform=ccrs.Geodetic())
ax.annotate('Beijing', xy=(beijing[0]+1, beijing[1]), color='black', fontsize=40)
```
